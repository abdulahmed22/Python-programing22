````python
import random
dice_rolls = [random.randint(1,6) for _ in range(20)]
with open('files/dice_rolls.txt', 'w') as dr:
    dr.write(f"Simulate 20 dice rolls:\n{dice_rolls}\n")
 
````

````python
dice_rolls.sort()
with open('files/dice_rolls.txt', 'a') as dr1:
    dr1.write(f"Sorted 20 dice rolls:\n{dice_rolls}\n")
    
````

````python
outcome = dice_rolls.count(4)
with open('files/dice_rolls.txt', 'a') as dr2:
    dr2.write(f"Number of fours:\n{outcome}\n")
````

````python
import random as rnd
from collections import Counter

def roll_dice(*dice, num_of_trials):
    counts = Counter()
    for roll in range(num_of_trials):
        counts[sum((rnd.randint(1,sides) for sides in dice))] +=1
    
    print(f"Number of rolls: {num_of_trials}")
    for outcome in range(len(dice), sum(dice)+1):
        print(f"{outcome}. {counts[outcome]}, probability: {(counts[outcome]/num_of_trials)}")

with open('files/simulation.txt', 'w') as fw:
    fw.write(roll_dice(6,num_of_trials=10) + '\n')
````
