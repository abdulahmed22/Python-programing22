````python
import matplotlib.pyplot as plt 

pichu_height = [21, 21, 20, 15, 10]
pichu_width = [28, 31, 15, 15, 10]
pikachu_height = [31, 51, 40, 35, 20]
pikachu_width = [25, 29, 40, 45, 30]


plt.plot(pichu_width, pichu_height, '*')
plt.plot(pikachu_width, pikachu_height, 'o')
plt.legend(["Pichu", "Pikachu"])
````

````python
import random as rnd
import matplotlib.pyplot as plt

pokemon = [
    [rnd.randint(20, 30) for _ in range(200)], # width
    [rnd.randint(15, 30) for _ in range(200)], # height
    [rnd.randint(0,1) for _ in range(200)], # label
]

plt.scatter(pokemon[0], pokemon[1], c = pokemon[2])
len(pokemon)
# width, height, c=type
````

````python
import numpy as np
pokemon_list2 = [
    [rnd.randint(20, 35), rnd.randint(20, 35), rnd.randint(0, 1)] for _ in range(200)
]

pokemon_array = np.array(pokemon_list2)

plt.scatter(pokemon_array[:,0],pokemon_array[:,1], c = pokemon_array[:,-1]  )
````
